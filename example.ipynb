
### Simple example on how to generate sine wave

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
time=np.arange(0, 10, 0.1)
amp=np.sin(time)
plt.plot(time, amp)
plt.title('Sine wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Display the sine wave

plt.show()
```
